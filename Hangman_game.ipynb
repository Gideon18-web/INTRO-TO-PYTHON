```python
import pandas as pd
import random

# List of possible words
word_list = ["python", "pandas", "data", "science", "analysis"]

# Randomly choose a word
chosen_word = random.choice(word_list)

# Create DataFrame with letters and visibility
df = pd.DataFrame({
    "letter": list(chosen_word),
    "revealed": [False] * len(chosen_word)
})

# Hangman stages (from 6 lives to 0)
hangman_stages = [
    """
     _______
     |     |
     |     O
     |    /|\\
     |    / \\
     |
    _|_
    """,
    """
     _______
     |     |
     |     O
     |    /|\\
     |    /
     |
    _|_
    """,
    """
     _______
     |     |
     |     O
     |    /|\\
     |
     |
    _|_
    """,
    """
     _______
     |     |
     |     O
     |    /|
     |
     |
    _|_
    """,
    """
     _______
     |     |
     |     O
     |     |
     |
     |
    _|_
    """,
    """
     _______
     |     |
     |     O
     |
     |
     |
    _|_
    """,
    """
     _______
     |     |
     |
     |
     |
     |
    _|_
    """
]

# Set initial game state
lives = 6
guessed_letters = []

print("üéØ Welcome to Hangman with Pandas!\n")

#Game loop
while lives > 0:
    # Display hangman drawing
    print(hangman_stages[6 - lives])

#Show current word with guessed letters
    display_word = df.apply(lambda row: row["letter"] if row["revealed"] else "_", axis=1)
    print("Word: ", " ".join(display_word))
    print(f"Lives left: {lives}")
    print("Guessed letters:", ", ".join(guessed_letters))

#Get user's guess
    guess = input("Guess a letter: ").lower()

    if guess in guessed_letters:
        print("‚ùó You already guessed that letter.\n")
        continue

    guessed_letters.append(guess)

    if guess in df["letter"].values:
        df.loc[df["letter"] == guess, "revealed"] = True
        print("‚úÖ Correct guess!\n")
    else:
        lives -= 1
        print("‚ùå Wrong guess. Part of the hangman is drawn.\n")

#Check for win
    if df["revealed"].all():
        print("üéâ Congratulations! You guessed the word:", chosen_word)
        break
else:
    print(hangman_stages[6])
    print("üíÄ Game over! You ran out of lives.")
    print("The word was:", chosen_word)
```
